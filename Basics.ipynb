### Traits

Traits are abstract data types containing certain fields and methods. <br>
In Scala inheritance, a class can only extend one other class, but it can extend multiple traits. <br>
**So traits allow for multple inheritance.**

```scala
trait Greeter {
  def greet(name: String): Unit
}
```

Traits can have default implementations.

```scala
trait Greeter {
  def greet(name: String): Unit =
    println("Hello, " + name + "!")
}
```

Example of inhertiance with traits.

* The `override` keyword is used to override the `greet` method.
* We can also override the constructure of the trait.

```scala
class DefaultGreeter extends Greeter

class CustomizableGreeter(prefix: String, postfix: String) extends Greeter {
  override def greet(name: String): Unit = {
    println(prefix + name + postfix)
  }
}
```

```scala
val greeter = new DefaultGreeter()
greeter.greet("Scala developer") // Hello, Scala developer!
```

```scala
val customGreeter = new CustomizableGreeter("How are you, ", "?")
customGreeter.greet("Scala developer") // How are you, Scala developer?
```

### Basic Encapsulation

The class definition below features private data: `_x`, `_y`, public getter methods `x` and `y`, and public setter methods `x_` and `y_`.

```scala
class Point {
  private var _x = 0
  private var _y = 0
  private val bound = 100

  def x = _x
  def x_= (newValue: Int): Unit = {
    if (newValue < bound) _x = newValue else printWarning
  }

  def y = _y
  def y_= (newValue: Int): Unit = {
    if (newValue < bound) _y = newValue else printWarning
  }

  private def printWarning = println("WARNING: Out of bounds")
}
```

`_x` and `_y` are vars, so can me mutated, like below.

```scala
val point1 = new Point
point1.x = 99
point1.y = 101 // prints the warning
```

However in this example, the following doesn't compile because vals are immutable.

```scala
class Point(val x: Int, val y: Int)
val point = new Point(1, 2)
point.x = 3  // <-- does not compile
```

Using val or var is good practice, however if don't state it, val is the default.

`private` is the default access modifier.

## Traits

Similar to a Java 8 interface.

```scala
trait HairColour
```

The trait below is called `Iterator`, it takes a _type_ `A`, and has methods: `hasNext` and `next`.

```scala
trait Iterator[A] {
    def hasNext: Boolean
    def next(): A
}
```

The class `IntIterator` below extends the `Iterator` trait above. 

It implements its abstract methods: `hasNext` and `next`.

```scala
class IntIterator(to: Int) extends Iterator[Int] {
  private var current = 0
  override def hasNext: Boolean = current < to
  override def next(): Int = {
    if (hasNext) {
      val t = current
      current += 1
      t
    } else 0
  }
}
```

Below we instantiate a new `iterator` with a `to` value of `10`.

```scala
val iterator = new IntIterator(10)
iterator.next()  // returns 0
iterator.next()  // returns 1
```

Calls to `.next()` will return increasing numbers up to `9` then it will go back to `0`.

## Mixins

Can use mixins to compose classes.

First we define an `abstract` class `A`.

```scala
abstract class A {
  val message: String
}
```

Then we define a class `B` which inherits from `A`.

```scala
class B extends A {
  val message = "I'm an instance of class B"
}
```

The trait `C` extends `A` and adds a method `loudMessage`.

```scala
trait C extends A {
  def loudMessage = message.toUpperCase()
}
```

Now we can make a class `D` just by extending `B` and adding the trait `C`.

```scala
class D extends B with C
```

We can now instantiate the class `D` and use its methods.

```scala
val d = new D
println(d.message)  // I'm an instance of class B
println(d.loudMessage)  // I'M AN INSTANCE OF CLASS B
```

A `class` can only have one `superclass` but many `traits` using the `with` keyword.

Another example shows us how we can mixin more fundamental methods with traits.

```scala
abstract class AbsIterator {
  type T // Define the type as a type
  def hasNext: Boolean
  def next(): T
}
```

Extend the `AbsIterator` to create `StringIterator`. <br>
This is a concrete class, which has an implementation of every abstract method.

```scala
class StringIterator(s: String) extends AbsIterator {
  type T = Char
  private var i = 0 // initialise counter
  def hasNext = i < s.length // is the counter below the length of the string
  def next() = {
    val ch = s charAt i // get the character at position i
    i += 1 // increment the counter
    ch // return the character
  }
}
```

We can now make a trait called `RichIterator`.

```scala
trait RichIterator extends AbsIterator {
  def foreach(f: T => Unit): Unit = while (hasNext) f(next())
}
```

The class `RichStringIter` extends `StringIterator`  and has a method from the `RichIterator` trait.

```scala
class RichStringIter extends StringIterator("Scala") with RichIterator
val richStringIter = new RichStringIter
```

We can now use the `foreach` method which came from the `RichIterator` trait.

```scala
richStringIter.foreach(println)
```

Creating a case class.

```scala
case class Fruit(name: String)
```

Instantiate immutable class instances.

```scala
val apple = Fruit("apple")
val orange = Fruit("orange")
val kiwi = Fruit("kiwi")
```

Make a list.

```scala
val fruitBasket = List(apple, orange, kiwi)
```

Define a function to filter the list.

```scala
def getApples(basket: List[Fruit]) = 
    for (fruit <- basket if fruit.name == "apple") 
    yield fruit
```

```scala
getApples(fruitBasket)
```

Generic higher order approach.

```scala
def getFruits(basket: List[Fruit], filterByFruit: Fruit => Boolean) =
    for (fruit <- basket if filterByFruit(fruit)) 
    yield fruit
```

```scala
def getApples(basket: List[Fruit]) = 
    getFruits(fruitBasket, (fruit: Fruit)=> fruit.name == "apple")
```

```scala
getApples(fruitBasket)
```
